### 2. Run Preprocessing Pipeline

```python
df = preprocessor.remove_duplicates()
```

Drops rows missing critical columns (review_text, rating, bank_name)

```python
df = preprocessor.handle_missing_data()
```

Converts review_date to proper YYYY-MM-DD.

```python
preprocessor.normalize_dates()
```

```python
df = preprocessor.clean_text()
```

```python
df = preprocessor.validate_ratings()
```
